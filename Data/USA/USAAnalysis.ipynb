```python
csv3['Deaths per 1000 People'] = csv3['Deaths per 1000 People'].interpolate(method='polynomial', order=2)
csv3.head()
```

```python
csv3.loc[csv3['date'] >= '1945-12-31']
```

```python
import math
for i in range(1, len(csv3)):
    if math.isnan(csv3.loc[i, 'Deaths Annual % Change']):
        csv3.loc[i, 'Deaths Annual % Change'] = (1 -  csv3.loc[i-1, 'Deaths per 1000 People'] / csv3.loc[i, 'Deaths per 1000 People']) * 100

```
